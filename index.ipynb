## Rolling Statistics 

Use the `.rolling()` function to find the rolling mean and rolling std with a window of 12 months. Plot the original curve along with the rolling mean and standard error.

```python
# This cell was copied from the solution at https://github.com/learn-co-curriculum/dsc-testing-for-trends-lab/tree/solution
#Determine rolling statistics
rollingMean = df.rolling(window=12, center=False).mean()
rollingStd = df.rolling(window=12, center=False).std()
```

```python
# This cell was copied from the solution at https://github.com/learn-co-curriculum/dsc-testing-for-trends-lab/tree/solution
#Plot rolling statistics
plt.figure(figsize=(17, 8))
plt.plot(df, label='original')
plt.plot(rollingMean, label='rolling mean')
plt.plot(rollingStd, label='rolling std')
plt.legend()
plt.title('Rolling mean and standard deviation')
plt.show();
```

Though the variation in standard deviation is small, the mean is increasing with time and thus, this is not a stationary series. 

## Dickey-Fuller Test 

Use the Dickey-Fuller Test to verify your visual result.

```python
df.columns
```

```python
# This cell was copied from the solution at https://github.com/learn-co-curriculum/dsc-testing-for-trends-lab/tree/solution
from statsmodels.tsa.stattools import adfuller

#Perform Dickey-Fuller test:
print('Results from Dickey-Fuller Test')
df_test = adfuller(df['#Passengers']) # initiate adfuller object

# Extract and display test results in a user friendly manner        # make a series to organize output
df_output = pd.Series(df_test[:4], index=['Test Statistic', 'p-value', '#lags used', 'Number of observations used'])
for key, value in df_test[4].items():  # get critical values stored in the objects 4th index
    df_output['Critical Value (%s)'%key] = value
print(df_output)
    
# Results of Dickey-Fuller Test:
# Test Statistic                   0.815369
# p-value                          0.991880
# #Lags Used                      13.000000
# Number of Observations Used    130.000000
# Critical Value (1%)             -3.481682
# Critical Value (5%)             -2.884042
# Critical Value (10%)            -2.578770
# dtype: float64
```
